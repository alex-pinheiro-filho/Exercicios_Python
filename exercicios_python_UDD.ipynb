```python
# Exercício 8 - Existe alguma diferença de cobrança para quem tem filhos? 
# Identifique se ter filhos parece acarretar numa cobrança maior olhando a média de charges para quem tem 0 filhos 
# e a média para quem tem 1 ou mais filhos
media_cobranca_sem_filhos = df[df['children'] == 0]['charges'].mean()
media_cobranca_com_filhos = df[df['children'] >= 1]['charges'].mean()
diferenca_media_cobranca = media_cobranca_com_filhos - media_cobranca_sem_filhos

print(f"Média de cobrança para quem não tem filhos: {media_cobranca_sem_filhos:.2f}")
print(f"Média de cobrança para quem tem 1 ou mais filhos: {media_cobranca_com_filhos:.2f}")
print(f"Diferença da média do valor de cobrança: {diferenca_media_cobranca:.2f}")
```

Olhando o valor das médias de cobrança e o valor da diferença das médias, chegamos a conclusão de que ter filhos acarreta em um valor de cobrança mais alto.

```python
# Exercício 9 - Para o homem, o BMI de 20 a 24,9 é considerado normal. 
# Acima disso, é considerado acima do peso, abaixo é considerado abaixo do peso. 
# Crie uma nova coluna, chamada "bmi_male" que receberá o texto "abaixo do normal" para BMI abaixo de 20, 
# "normal" para BMI de 20 a 24,9 e "acima do normal" para BMI acima de 24,9.
import numpy as np

df['bmi_male'] = np.where(
    df['bmi'] < 20, "abaixo do normal",
    np.where(
        (df['bmi'] >= 20) & (df['bmi'] <= 24.9), "normal",
        "acima do normal"
    )
)

df.head(30)
```

```python
# Exercício 10 - Se uma pessoa é fumante e ainda tem um BMI acima do considerado normal, é possível que isso seja um indicador 
# negativo para a seguradora. Quantas pessas estão nessa situação?
pessoas_fumantes_bmi_alto = df[(df['smoker'] == 'yes') & (df['bmi'] > 24.9)]
contagem_pessoas_indicador_negativo = len(pessoas_fumantes_bmi_alto)

print(f"{contagem_pessoas_indicador_negativo} pessoas estão na situação de um possível indicador negativo para a seguradora.")
```
